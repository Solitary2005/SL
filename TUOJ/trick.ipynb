## 自定义排序

```python
students = [('Alice', 90, 20), ('Bob', 90, 19), ('Charlie', 80, 22)]
# key 返回一个元组，加负号表示降序
students.sort(key=lambda x: (-x[1], x[2], x[0]))
# 排完：Bob(90, 19) -> Alice(90, 20) -> Charlie(80, 22)
# 第一优先级：-x[1]（降序）
# 假设 x[1] 是成绩。由于 Python 默认是升序排列（从小到大），给数值取负号可以巧妙地实现降序。数值越大，取负后越小，就会排在前面。

# 第二优先级：x[2]（升序）
# 如果成绩（x[1]）相同，Python 会对比生成的第二个元素 x[2]。这里没取负号，所以是正常的升序排列。

# 第三优先级：x[0]（升序）
# 如果前两个字段都完全一样，则按 x[0]（通常是名字或学号）的字典序升序排列。
```

### BFS(广搜)

```python
# 广搜的搜索方式适合于解决两个点之间的最短路径问题
# 广搜是从起点出发，以起始点为中心一圈一圈进行搜索
# 一旦遇到终点，记录之前走过的节点就是一条最短路
# 通常用于求无权图/网格的最短路径（例如走迷宫）

target_x = 0
target_y = 0
grid = [[0]*n for _ in range(m)]

from collections import deque

def bfs(start_x, start_y):
    # 队列里通常存元组 (x, y, 当前步数)
    q = deque([(start_x, start_y, 0)]) 
    # visited 集合用于去重，避免重复访问导致死循环
    visited = set([(start_x, start_y)])
    # 也就是常说的方向数组：右、下、左、上
    directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
    
    while q:
        x, y, step = q.popleft()
        
        # 1. 检查是否到达终点
        if x == target_x and y == target_y:
            return step
            
        # 2. 遍历四个方向
        for dx, dy in directions:
            nx, ny = x + dx, y + dy
            
            # 3. 判断越界及是否可走（需根据具体题目修改） 这里 '#' 表示有障碍物
            if 0 <= nx < n and 0 <= ny < m and grid[nx][ny] != '#' and (nx, ny) not in visited:
                visited.add((nx, ny))
                q.append((nx, ny, step + 1))
                
    return -1 # 走不到终点
```

### 前缀和+差分数组

```python
# 经典问题：区间覆盖
# 问题特征：有n个区间，m个查询，求对于每个查询，有多少个区间覆盖了该查询
# 前缀和通常用来快速求一个数组的某个区间和：
# 对于数组a，希望求a[l]到a[r]的和，可以先对a进行前缀和求解，即pre[i] = pre[0]+...+pre[i], 则a[l]到a[r]的和为pre[r]-pre[l-1]

# 差分数组可用于快速给一个数组的某个区间统一加上一个值：
# 即 有一个区间，我们想给区间 [L, R] 每个位置加一个常数scaler

# --- 前缀和 ---
# 原数组 a: [1, 2, 3, 4]
# pre[i] 表示 a[0] 到 a[i-1] 的和
a = [1, 2, 3, 4]
pre = [0] * (len(a) + 1)
for i in range(1, len(a) + 1):
    pre[i] = pre[i-1] + a[i-1]
# 查询区间 [L, R] 的和 (0-indexed)：
# sum_LR = pre[R+1] - pre[L]

# --- 差分数组 ---
# 对区间 [L, R] 全部加上 val
# 原数组 a: [0, 0, 0, 0, 0]
n = 5
diff = [0] * (n + 1) # 多开一位防止越界

def add(L, R, val):
    diff[L] += val
    diff[R+1] -= val

# 执行几次操作
add(1, 3, 5)  # 区间 [1, 3] + 5
add(2, 4, 2)  # 区间 [2, 4] + 2

# 还原出最终的数组 (差分数组求前缀和即为原数组)
res = [0] * n
res[0] = diff[0]
for i in range(1, n):
    res[i] = res[i-1] + diff[i]
print(res) # [0, 5, 7, 7, 2]
```

### 快速分解质因数

```python
# 质因数分解快速方法：试除法
# 从2开始试除，如果j是n的因子，就把j除干净; 
# 只用枚举到根号n，因为大于根号n的因子的指数一定是1（p>根号n, 则p^2>n, 所以必然只能为1）

def prime_factors(n):
    factors = []
    d = 2
    # 只需要枚举到根号 n
    while d * d <= n:
        if n % d == 0:
            count = 0
            while n % d == 0:
                count += 1
                n //= d
            factors.append((d, count)) # 记录质因子及其指数
        d += 1
    # 如果最后 n 还没除尽，说明剩下的是一个大于根号 n 的质数
    if n > 1:
        factors.append((n, 1))
    return factors

print(prime_factors(360)) # [(2, 3), (3, 2), (5, 1)]
```

### 求解区间内所有质数

```python
import math

def segmented_sieve(a, b):
    """
    返回区间 [a, b] 内的所有质数（包含端点）。
    """
    if b < 2:
        return []
    if a < 2:
        a = 2

    limit = int(math.isqrt(b)) + 1
    # 生成 [2, sqrt(b)] 范围内的质数（基础筛）
    is_prime_small = [True] * (limit + 1)
    is_prime_small[0] = is_prime_small[1] = False
    for i in range(2, int(math.isqrt(limit)) + 1):
        if is_prime_small[i]:
            for j in range(i * i, limit + 1, i):
                is_prime_small[j] = False
    primes = [i for i in range(2, limit + 1) if is_prime_small[i]]

    # 分段筛
    segment_size = int(math.isqrt(b - a + 1)) + 1  # 分块大小
    result = []
    # 对每个块进行筛选
    for low in range(a, b + 1, segment_size):
        high = min(low + segment_size - 1, b)
        is_prime_seg = [True] * (high - low + 1)

        for p in primes:
            if p * p > high:
                break
            # 找到第一个 >= low 的 p 的倍数
            start = max(p * p, ((low + p - 1) // p) * p)
            for j in range(start, high + 1, p):
                is_prime_seg[j - low] = False

        for i in range(low, high + 1):
            if is_prime_seg[i - low]:
                result.append(i)

    return result

# 示例用法
if __name__ == "__main__":
    a, b = 10, 50
    primes = segmented_sieve(a, b)
    print(f"区间 [{a}, {b}] 内的质数有: {primes}")
```

### 进制转换

快速取位与组合取低 $k$ 位：n & ((1 << k) - 1)

判断第 $i$ 位是否为 1（从 0 开始计数）：(n >> i) & 1

```python
# ================================================= =
# ## 位运算、进制转换与计算机底层表示 (补码/移码)
# 取模与位运算：在对 $2^k$ 取模时，n % (2**k) 等价于 n & ((1 << k) - 1)，后者效率更高。
# 符号位陷阱：Python 的 bin(-1) 结果是 '-0b1'。在 CSP 题目要求输出“8位补码”时，必须用 bin(-1 & 0xFF) 来强制获得 0b11111111。
# 大数加法：如果题目涉及大整数运算，Python 不需要像 C++ 那样写高精度算法，直接加减即可，
# 但在最后输出结果时，如果题目限定了机器字长（如 32 位），记得用 & 0xFFFFFFFF 截断。
# ================================================= =

import math

# --- 1. 基础位运算 Trick ---
n = 0b1010  # 10进制中的 10
# 判断第 i 位是否为 1 (i从0开始)
is_one = (n >> 1) & 1  # 结果为 1
# 将第 i 位设为 1
n_set = n | (1 << 2)   # 1010 | 0100 = 1110 (14)
# 快速判断 2 的幂次 (2, 4, 8, 16...)
is_pow2 = n > 0 and (n & (n - 1)) == 0
# lowbit 运算 (树状数组核心): 返回最低位的 1 及其后面的 0
def lowbit(x): return x & -x

# --- 2. 进制转换 ---
# 10进制转其他 (自带函数)
b_str = bin(10)  # '0b1010'
o_str = oct(10)  # '0o12'
h_str = hex(255) # '0xff'
# 其他进制转 10 进制
num = int("ff", 16)  # 255
num = int("1010", 2) # 10

# 手动转指定位数的二进制字符串 (补齐前导0)
def to_bin_fixed(n, bits=8):
    return bin(n & ((1 << bits) - 1))[2:].zfill(bits)

# --- 3. 补码 (Two's Complement) 处理 ---
# Python 整数是无限精度的，模拟 C++ 32位整数溢出需要手动截断
def to_32bit_int(n):
    n = n & 0xFFFFFFFF        # 核心：取低32位
    if n > 0x7FFFFFFF:        # 如果最高位(符号位)为1
        n -= 0x100000000      # 转换为 Python 的负数表示
    return n

# 计算 n 的 bits 位补码表示
def get_complement(n, bits=8):
    # 如果 n 是负数，n & mask 会自动计算出其在该位宽下的补码
    mask = (1 << bits) - 1
    return bin(n & mask)[2:].zfill(bits)

# --- 4. 移码 (Excess-K) 处理 ---
# 移码 = 真值 + 偏置 (通常偏置为 2^(n-1))
# 技巧：n 位移码通常等于 n 位补码的符号位取反
def get_excess_code(n, bits=8):
    complement_val = n & ((1 << bits) - 1)
    excess_val = complement_val ^ (1 << (bits - 1)) # 符号位取反
    return bin(excess_val)[2:].zfill(bits)

# --- 示例测试 ---
# -1 的 8 位补码: 11111111
# -1 的 8 位移码: 01111111 (符号位从1变0)
print(f"补码: {get_complement(-1, 8)}")
print(f"移码: {get_excess_code(-1, 8)}")
```

### 暴力枚举

```python
import itertools

# 全排列
items = [1, 2, 3]
for p in itertools.permutations(items):
    print(p) # (1, 2, 3), (1, 3, 2) ...

# 组合（从 4 个里选 2 个）
for c in itertools.combinations([1, 2, 3, 4], 2):
    print(c) # (1, 2), (1, 3), (1, 4), (2, 3) ...
```

### Dijkstra 最短路径

```python
# 场景：带权图的最短路径
# 假设有 n 个节点，给定边和权重，求从起点 start 到所有点的最短距离。

import heapq
from collections import defaultdict

def dijkstra(n, edges, start):
    # 1. 建图：使用邻接表
    # graph[u] = [(v, weight), ...]
    graph = defaultdict(list)
    for u, v, w in edges:
        graph[u].append((v, w))
    
    # 2. 初始化距离数组，初始值为无穷大
    # dists[i] 表示从 start 到 i 的当前最短距离
    dists = [float('inf')] * (n + 1)
    dists[start] = 0
    
    # 3. 初始化堆 (pq: priority queue)
    # 堆中存放元组: (当前累计距离, 当前节点编号)
    # 注意：heapq 默认是小根堆，会根据元组的第一个元素（距离）自动排序
    pq = [(0, start)]
    
    while pq:
        # 弹出当前距离起点最近的节点
        curr_dist, u = heapq.heappop(pq)
        
        # 4. 核心优化：如果弹出的距离已经大于已知最短距离，说明是失效的过时信息，跳过
        if curr_dist > dists[u]:
            continue
            
        # 5. 遍历当前节点的所有邻居
        for v, weight in graph[u]:
            new_dist = curr_dist + weight
            
            # 如果发现更短的路径，则更新并入堆
            if new_dist < dists[v]:
                dists[v] = new_dist
                heapq.heappush(pq, (new_dist, v))
                
    return dists

# --- 测试数据 ---
# 节点 1-4，边 (u, v, w)
n = 4
edges = [
    (1, 2, 5), (1, 3, 2), 
    (3, 2, 1), (2, 4, 3), (3, 4, 10)
]
print(dijkstra(n, edges, 1)) 
# 输出: [inf, 0, 3, 2, 6] (第0位不用，1->2最短是3, 1->3是2, 1->4是6)
```

### 贪心

```python
# heapq 在贪心题中也常用于动态维护最大/最小值
# 小根堆：合并果子，每次弹出两个最小的合并，再把结果放回去。

import heapq
# 假设 piles = [1, 2, 9]
heapq.heapify(piles) # 将列表原地转化为堆 O(n)
total_cost = 0
while len(piles) > 1:
    combined = heapq.heappop(piles) + heapq.heappop(piles)
    total_cost += combined
    heapq.heappush(piles, combined)
```

```python
# 大根堆
# 动态维护最大值
# 技巧：取负数存入，弹出后再取负回来
import heapq

nums = [1, 2, 9]
max_heap = []

for x in nums:
    # 存入负数：-9, -2, -1
    # 在小根堆里，-9 是最小值，排在堆顶
    heapq.heappush(max_heap, -x)

# 弹出时
max_val = -heapq.heappop(max_heap) 
print(max_val) # 输出 9，成功模拟了大根堆
```
