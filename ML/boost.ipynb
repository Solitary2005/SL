# XGBoost

```python
import xgboost as xgb

# 基础参数（重点控制过拟合）
params = {
    'objective': 'reg:squarederror',
    'learning_rate': 0.05,          # 较低的学习率，配合更多树
    'max_depth': 5,                 # 限制树深度防止过拟合
    'min_child_weight': 5,          # 叶子节点最小权重和
    'gamma': 0.1,                  # 分裂所需的最小损失减少
    'subsample': 0.8,              # 行采样比例
    'colsample_bytree': 0.8,       # 列采样比例
    'reg_alpha': 0.1,              # L1 正则化
    'reg_lambda': 1.0,             # L2 正则化
    'n_estimators': 2000,          # 最大树数量，由早停决定实际使用数
    'random_state': 42,
    'verbosity': 0
}

model = xgb.XGBRegressor(**params)
```
