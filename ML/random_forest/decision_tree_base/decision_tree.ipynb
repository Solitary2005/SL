# 决策树基础

```python
# sklearn 构建，训练，剪枝 决策树
from sklearn.tree import DecisionTreeClassifier

# 剪枝
# 后剪枝
def get_ccp_path(clf: DecisionTreeClassifier, X_train: pd.DataFrame, y_train: pd.Series) -> np.ndarray:
    '''
    参数：
        clf：预训练好的决策树分类器模型。
        X_train：训练集的特征数据。
        y_train：训练集的标签数据。
    功能：
        提取决策树模型中剪枝路径的相关信息，基于训练数据计算所有可能的剪枝参数。
    返回值：
        一个表示剪枝路径参数的数组，用于指导剪枝操作。
    '''
    path = clf.cost_complexity_pruning_path(X_train, y_train)
    return path.ccp_alphas

def train_pruned_models(X_train: pd.DataFrame, y_train: pd.Series, clf: DecisionTreeClassifier) -> Dict[float, DecisionTreeClassifier]:
    '''
    参数：
        X_train：训练集的特征数据。
        y_train：训练集的标签数据。
        clf：预训练的决策树分类器模型，用于获取初始参数和剪枝路径。
    功能：
        使用从 get_ccp_path 获取的剪枝路径参数。
        基于初始模型的参数，依次训练多个经过剪枝的模型，每个模型对应不同的剪枝参数值。
        将训练好的模型存储到字典中。
    返回值：
        一个字典，其中键为剪枝路径参数，值为对应剪枝后训练好的决策树模型。
    '''
    alphas = get_ccp_path(clf, X_train, y_train)
    res = {}
    from sklearn.base import clone
    for a in alphas:
        clf_a = clone(clf)
        clf_a.set_params(ccp_alpha=a)
        clf_a.fit(X_train, y_train)
        res[a] = clf_a
    return res


```
